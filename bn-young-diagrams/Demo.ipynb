# Demo of bnyoungdiagrams.py
First, we import the functions that compute the most probable diagram, plot rotated diagrams and limit shapes. 

```python
from bnyoungdiagrams import findmax, plot_rotated_diagram, plot_limit_shape, rho
import numpy as np
import matplotlib.pyplot as plt
np.seterr(all='ignore')
```

Let's compute the most probable diagram for the tensor power N=20 of the spinor representation of so(9). n=4

```python
a=findmax(4,20)
a
```

Plot the diagram

```python
plot_rotated_diagram(a)
```

Let's do it for N=50, n=13 (so(27))

```python
plot_rotated_diagram(findmax(13,50))
```

Now let us plot the limit shape for c=6, take n=10, N=41, c=(N+2n-1)/n=6

```python
plot_limit_shape(10,41)
```

Compare the limit shape to the diagram

```python
plot_rotated_diagram(findmax(10,41))
plot_limit_shape(10,41)
```

Now take n=30, N=121, c is again 6

```python
plot_rotated_diagram(findmax(30,121))
plot_limit_shape(10,41)
```

Plot limit shapes for the different values of c: 2,3,4,6,15 and a diagram for n=10, N=21

```python
plot_limit_shape(10,1)
plot_limit_shape(10,11)
plot_limit_shape(10,21)
plot_limit_shape(10,41)
plot_limit_shape(10,131)
plot_rotated_diagram(findmax(10,21))
```

We can also plot the densities for various values of c as in Fig 5 in the paper


```python
x=np.arange(-4,4,1/100)
plt.plot(x,rho(x,2.1))
plt.plot(x,rho(x,3))
plt.plot(x,rho(x,4))
plt.plot(x,rho(x,8))
```
